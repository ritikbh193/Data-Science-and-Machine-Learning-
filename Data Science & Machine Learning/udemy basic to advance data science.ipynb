```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.datasets import load_digits
```

```python
digit = load_digits()
```

```python
dir(digit)
```

```python
digit['target']
```

```python
for i in range(1,9):  
    plt.matshow(digit['images'][i])
    plt.show()
```

```python
x = digit['images']
y=digit['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
x_train1 = x_train.reshape(len(x_train),8*8)
```

```python
x_test1 = x_test.reshape(len(x_test),8*8)
```

```python
x_train1.shape
```

```python
model = LinearRegression()
```

```python
model.fit(x_train1,y_train)
```

```python
model.score(x_test1,y_test)
```

```python
x_train2 = x_train1/16
x_test2 = x_test1/16
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression()
```

```python
model1.fit(x_train2,y_train)
```

```python
model1.score(x_test2,y_test)
```

```python
y_pred = model1.predict(x_test2)
```
